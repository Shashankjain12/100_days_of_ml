```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
dataset=pd.read_csv("data.csv",sep=";")
```

```python
dataset.head()
dataset.columns
dataset.columns=["fixed_acidity","volatile_acidity","citric_acid","residual_sugar","chlorides","free_sulfur_dioxide","total_sulfur_dioxide","density","pH","sulphates","alcohol","quality"]
```

```python
cols_to_norm=["fixed_acidity","volatile_acidity","citric_acid","residual_sugar","chlorides","free_sulfur_dioxide","total_sulfur_dioxide","density","pH","sulphates","alcohol"]
dataset[cols_to_norm]=dataset[cols_to_norm].apply(lambda x:(x-x.min())/(x.max()-x.min()))
X=dataset.drop("quality",axis=1)
y=dataset["quality"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train.head()
```

```python
dataset.head()
dataset.columns
```

```python
fixed_acidity=tf.feature_column.numeric_column("fixed_acidity")
volatile_acidity=tf.feature_column.numeric_column("volatile_acidity")
citric_acid=tf.feature_column.numeric_column("citric_acid")
residual_sugar=tf.feature_column.numeric_column("residual_sugar")
chlorides=tf.feature_column.numeric_column("chlorides")
free_sulfur_dioxide=tf.feature_column.numeric_column("free_sulfur_dioxide")
total_sulfur_dioxide=tf.feature_column.numeric_column("total_sulfur_dioxide")
density=tf.feature_column.numeric_column("density")
pH=tf.feature_column.numeric_column("pH")
sulphates=tf.feature_column.numeric_column("sulphates")
alcohol=tf.feature_column.numeric_column("alcohol")
quality=tf.feature_column.numeric_column("quality")
```

```python
input_fun= tf.estimator.inputs.pandas_input_fn(x=X_train,y=y_train,batch_size=100,num_epochs=10,shuffle=True)
```

```python
train_fn= tf.estimator.inputs.pandas_input_fn(x=X_train,y=y_train,batch_size=100,num_epochs=10,shuffle=True)
```

```python
eval_input_fn= tf.estimator.inputs.pandas_input_fn(x=X_test,y=y_test,batch_size=100,num_epochs=10,shuffle=False)
```

```python
feat_cols=[fixed_acidity,volatile_acidity,citric_acid,residual_sugar,chlorides,free_sulfur_dioxide,total_sulfur_dioxide,density,pH,sulphates,alcohol]
model=tf.estimator.LinearRegressor(feature_columns=feat_cols)
```

```python
model.train(input_fn=input_fun,steps=100)
```

```python
model.evaluate(train_fn,steps=1000)
```

```python
list(model.predict(eval_input_fn))
```
